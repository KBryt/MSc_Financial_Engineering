## **Linear Discriminant Analysis**
**Basics:**

In the case of classification problems, supervised learning algorithms like LDA are applied. This approach to dimensionality reduction seeks a linear combination of features that separates the various classes of the dataset with maximum distance. LDA assumes that the covariance matrices of the different courses are identical and that data is Gaussian distributed. To deal better with class separation, LDA projects the original feature space into a lower-dimensional space.

**Classification**

The main categorization of LDAs is, however, their classification according to the type of tasks performed, such as;

1. Kernel Discriminant Analysis: This is the type of LDA in the kernel functions are involved in translating instances of input data to some higher-dimensional feature space.
2. Fisher's Linear Discriminant Analysis: This happens to be such an LDA variable that linearly combines the characteristics to maximize the between-class separation and to minimize the within-class variation.
3. Quadratic Discriminant Analysis: In this method, there can be a different covariance matrix for each class. In contrast, sparse discriminant analysis reduces the features by selecting only a subset of features.
4. Sparse discriminant analysis, on the other hand, reduces the features using feature selection.
5. Regularized DA: The methodology penalizes in the form of fines for overfitting the data.

### **Advantages:**

### **Disadvantages:**
1. **Linearity assumption:** LDA assumes that the decision boundary between classes is linear. If the underlying relationship in the time series data is non-linear, LDA may not perform well.
2. **Limited to classification:** LDA is primarily used for classification tasks and may not be suitable for other types of time series analysis, such as forecasting or regression.



### **Equations:**
Within-Class Scatter Matrix
$$ S_W = \sum_{i=1}^{c} \sum_{x \in X_i} (x - \mu_i)(x - \mu_i)^T $$

Between-class scatter matrix
$$ S_B = \sum_{i=1}^{c} N_i (\mu_i - \mu)(\mu_i - \mu)^T $$

Fisher's criterion
$$ J(w) = \frac{w^T S_B w}{w^T S_W w} $$



### **Features:**
1. **Works well with missing values:** LDA can handle missing values in the time series data by estimating the missing values based on the available data.
2. **Feature extraction**: LDA not only performs classification but also provides insights into the most discriminative features contributing to the separation between classes.

### **Hyperparameters:**
1. **Solver:** The solver algorithm used to compute the discriminant vectors (e.g., "svd" for singular value decomposition, "eigen" for eigenvalue decomposition).
2. **Shrinkage:** A parameter that controls the degree of shrinkage applied to covariance estimation.

## **Support Vector Machine**

### **Advantages:**

Support Vector Machines (SVM) offer several benefits when applied to time series analysis:
1. **Effective in high-dimensional spaces:** SVM can handle high-dimensional time series data and perform well even when the number of features is greater than the number of samples. This makes SVM suitable for analyzing time series with a large number of variables.
2. **Non-linear classification:** SVM can efficiently handle non-linear relationships in time series data by using kernel functions to map the data into a higher-dimensional space where linear separation is possible.
3. **Robust to overfitting:** SVM aims to find the optimal hyperplane that maximizes the margin between classes, which helps in generalizing well to unseen data and reduces the risk of overfitting.
4. **Versatile:** SVM can be used for both classification and regression tasks, making it applicable to various time series analysis problems.

### **Equations:**


Separating Hyperplane

In a p-dimensional space, the equation of a hyperplane is:

$\beta_{0} + \beta_{1}X_{1} + \beta_{2}X_{2} + \cdots +\beta_{p}X_{p} = 0$


Maximal Margin Classifier


$\sum_{j=1}^{p}\beta_{j}^{2} = 1$

Using class labels yi as -1 or 1, a correct classification is given by:

$y_{i} \left( \sum_{j=1}^{p}\beta_{j}X_{ij} + \beta_{0} \right) > 0$

Misclassification is defined as:

$y_{i} \left( \sum_{j=1}^{p}\beta_{j}X_{ij} + \beta_{0} \right) \leq 0$

Inner Product and Kernels

The inner product for two vectors a and b is:

$\left<a, b\right> = \sum_{i=1}^{n} a_i b_i$

For a new point X, a linear support vector machine is:

$y(x) = \beta_0 + \sum_{i}^{n} \alpha_{i} \left<X, X_{i} \right>$


The Gaussian Radial Basis Function kernel is given by:

$K(X_{i}, X_{i'}) = \text{exp} \left( -\gamma ||X_{i} - X_{i'}||^2 \right)$




### **Features:**

# **Step 4: Technical Section: Hyperparameter Tuning**








One of the essential steps that go into bettering the performance of ML models has to do with hyperparameter tuning. In other words, hyperparameters are those that are predefined and can't be learned from data-driven analysis. They regulate the model's behavior and complexity, and making adjustments to them can have a big effect on how well the model performs when applied to time series data. This section will cover the process of fine-tuning hyperparameters and include instances of hyperparameters used for neural networks, support vector machines (SVM), and linear discriminant analysis (LDA).

**Hyperparameter Tuning Process:**
1. Describe the area under search: Ascertain the range or range of values that are possible for every hyperparameter that you wish to adjust.

2. Choose a search approach: Select a technique, such as grid search, random search, or Bayesian optimization, to investigate the hyperparameter search space.
3. Evaluation Metric: Define some metric by which the model is to be evaluated for its performance about different hyperparameter settings applied. Depending on the task, this might be mean squared error, recall, accuracy, precision, or any other relevant statistic.

4. Cross-validation: Since the data is time series, the time-series data will be split into training and validation sets. Walk-forward validation or rolling-window cross-validation can ensure that a realistic assessment of model performance does occur during the evaluation.
5. Hyperparameter Tuning: Run the chosen search algorithm to discover the best combination of hyperparameters for optimizing the assessment metric. Investigate model performance on the validation set for different settings of the hyperparameters.

6. Model selection: According to your evaluation metric, choose the best-performing model on the validation set.
7. Examine on a test set: Following the selection of the optimal model, determine the model's generalization capabilities by evaluating its performance on a separate test set.

Examples of Hyperparameters for Time Series Models:

1. **Linear Discriminant Analysis (LDA):**
- **Solver:** There are three options: 'eigen' (Eigenvalue Decomposition), 'lsqr' (Least Squares), and'svd' (Singular Value Decomposition).
- **Shrinkage:** The covariance matrix is estimated using the shrinkage parameter. It regulates the ratio of a priori knowledge to the sample covariance matrix.
- **n_components:** The quantity of discriminant components to retain, n_components. It establishes how many dimensions the modified feature space has.

2. **Support Vector Machines (SVM):**
- **Kernel:** The kernel function can be linear, polynomial, sigmoid, or the radial basis function (RBF). Hyperparameters are unique to each kernel.
- **C:** Achieving a low training error and a low complexity model are traded off, and this is managed by the regularization parameter C.
- **Gamma:** The 'rbf', 'poly', and'sigmoid' kernel coefficients. It determines the impact of every training case and influences the decision boundary's degree of flexibility.

3. **Neural Networks:**
- **Number of layers:** The count of concealed layers within the architecture of a neural network.
- **Number of neurons:** The quantity of neurons found in every hidden layer.
- **Learning rate:** The speed at which weights are updated by the model while it is being trained.
- **Batch size:** The quantity of samples used in a training cycle.
- **Activation functions:**  The sigmoid, tanh, or ReLU activation function that each neuron chooses to use.
- **Regularization parameters:** The hyperparameters that regulate regularization methods such as $L_1/L_2$ regularization or dropout regularization.

**Note:** The particular hyperparameters and their ranges may vary based on the library and implementation (e.g., scikit-learn, TensorFlow, PyTorch). For comprehensive information about hyperparameters, it is crucial to refer to the documentation of the particular library.

You can determine the ideal configuration that maximizes the model's performance on your time series data by methodically adjusting these hyperparameters.
