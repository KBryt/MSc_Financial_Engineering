Ridge regression is a linear regression technique that addresses the problem of multicollinearity (high correlation) among predictor variables by introducing $L_2$ regularization. ##**Team Member B: k-means Clustering**
### **Basics:**

K-means clustering is an unsupervised machine learning algorithm used for partitioning a dataset into a predetermined number of clusters. It aims to group similar data points together based on their feature similarity. The algorithm iteratively assigns each data point to the nearest centroid (representative point of a cluster) and updates the centroids by calculating the mean of the assigned data points. This process continues until convergence, where the centroids stabilize and the assignments remain unchanged.

K-means clustering is a centroid-based clustering technique that minimizes the variance within each cluster. It assigns data points to clusters based on the proximity to the centroids, which represent the center of each cluster. The algorithm's goal is to find the optimal arrangement of clusters that minimizes the overall variance within the data.

###**Classification:**

Partitioning techniques in unsupervised learning include K-means clustering. Since it uses the idea of centroids, also known as cluster centers, to identify clusters in the data, it is referred to as a centroid-based clustering method. Since K-means represents each cluster by its centroid, which acts as a prototype for the cluster, it may also be classified as a prototype-based clustering technique.

```python
import numpy as np
import pandas as pd
import yfinance as yf
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split

# Getting historical market data from SPY (ETF) (SPY)
df = yf.download("SPY", start="2000-01-01", end="2022-01-01")

# Calculate returns and create lagged features
df["Ret"] = df["Adj Close"].pct_change()
name = "Ret"
df["Ret10_i"] = (
    df[name].rolling(10).apply(lambda x: 100 * ((np.prod(1 + x)) ** (1 / 10) - 1))
)
df["Ret25_i"] = (
    df[name].rolling(25).apply(lambda x: 100 * ((np.prod(1 + x)) ** (1 / 25) - 1))
)
df["Ret60_i"] = (
    df[name].rolling(60).apply(lambda x: 100 * ((np.prod(1 + x)) ** (1 / 60) - 1))
)
df["Ret120_i"] = (
    df[name].rolling(120).apply(lambda x: 100 * ((np.prod(1 + x)) ** (1 / 120) - 1))
)
df["Ret240_i"] = (
    df[name].rolling(240).apply(lambda x: 100 * ((np.prod(1 + x)) ** (1 / 240) - 1))
)

# Remove unnecessary columns and drop NA values
del df["Open"]
del df["Close"]
del df["High"]
del df["Low"]
del df["Volume"]
del df["Adj Close"]
df = df.dropna()

# Shift Ret25_i by -25 to create target variable and drop NA values
df["Ret25"] = df["Ret25_i"].shift(-25)
df = df.dropna()

# Define input matrix X and target vector y
X, y = df.iloc[:, 0:-1], df.iloc[:, -1]
print(X.shape, y.shape)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=int(len(y) * 0.5), shuffle=False
)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

# Ridge Regression
ridgeR = Ridge(alpha=0.7)
ridgeR.fit(X_train, y_train)
y_pred_ridge = ridgeR.predict(X_train)
y_pred_test_ridge = ridgeR.predict(X_test)

# Calculate mean squared error on test set
mean_squared_error = np.mean((y_pred_test_ridge - y_test) ** 2)
print("Mean squared error on test set", mean_squared_error)

# Get and plot ridge coefficients
ridge_coefficient = pd.DataFrame()
ridge_coefficient["Columns"] = X_train.columns
ridge_coefficient["Coefficient Estimate"] = pd.Series(ridgeR.coef_)
print(ridge_coefficient)

fig, ax = plt.subplots(figsize=(15, 7))
color = [
    "tab:gray",
    "tab:blue",
    "tab:orange",
    "tab:green",
    "tab:red",
    "tab:purple",
    "tab:brown",
    "tab:pink",
    "tab:gray",
    "tab:olive",
    "tab:cyan",
    "tab:orange",
    "tab:green",
    "tab:blue",
    "tab:olive",
]
ax.bar(
    ridge_coefficient["Columns"],
    ridge_coefficient["Coefficient Estimate"],
    color=color,
)

plt.style.use("ggplot")
plt.show()


#######################
```

### **Equations**

The regularization term is of the form
$$
\begin{align*}
\alpha\sum_{i=1}^n \theta_i^2
\end{align*}
$$
All parameters will be near to zero if $\alpha$ is very large; if $\alpha=0$, we can easily not regularize the model.  

This gives the cost function for a linear regression with a ridge penalty (Ridge Regression)
$$
\begin{align*}
J(\theta) = MSE(\theta) + \frac{1}{2}\alpha\sum_{i=1}^n \theta_i^2
\end{align*}
$$

Gradient descent, therefore, will result in the following update of the parameters
$$
\begin{align*}
\theta^{[next\ step]}=\theta^{[previous\ step]} - \eta \nabla_{\theta} MSE\left(\theta^{[previous\ step]}\right) - a \odot \theta^{[previous\ step]}
\end{align*}
$$

where $a$ is a column vector made up of $n$ items, all of which, with the exception of the first term, have a value of $\alpha$.

### **Computation**

The code demonstrates an example of k-means clustering using the scikit-learn library. Here's a summary of what the code accomplishes in a few paragraphs:

- The necessary libraries are imported, including `matplotlib.pyplot` for data visualization, `numpy` for numerical operations, `sklearn.cluster.KMeans` for k-means clustering, and `sklearn.datasets.make_blobs` to generate a synthetic dataset with clusters.

- A random seed is set for reproducibility. Then, a synthetic dataset with 4 clusters (`centers=4`) and 2500 samples is generated using the `make_blobs` function. The input samples are stored in `X`, and the corresponding cluster labels are stored in `y`.

- The elbow curve is plotted to determine the optimal number of clusters (`K`) for the k-means algorithm. The within-cluster sum of squares (WCSS) is calculated for different values of `K` by fitting the k-means model to the dataset using a loop. The WCSS values are stored in the `wcss` list. The plot shows the number of clusters on the x-axis and the corresponding WCSS on the y-axis.

- K-means clustering is performed using the optimal number of clusters (`n_clusters=4`) determined from the elbow curve. The k-means model is fitted to the dataset using `fit_predict`, which assigns each sample to a cluster based on their features. The cluster assignments are stored in the `ymeans` variable.

- The clustering results are visualized by creating a scatter plot. Each cluster is represented by a different color, and the cluster centers are marked with black crosses. This example serves as a demonstration of the k-means clustering algorithm and its application in grouping similar data points into distinct clusters.

```python
# k-means Clustering Example
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

np.random.seed(0)

# Generate dataset
X, y = make_blobs(centers=4, n_samples=2500)

# Plot dataset
fig = plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.title("Dataset with 4 clusters", fontsize=20)
plt.xlabel("First feature")
plt.ylabel("Second feature")
plt.suptitle("Fig. 5: Dataset Visualization", fontweight="bold", horizontalalignment="right")
plt.show()

# Plot elbow curve
wcss = []
for i in range(1, 11):
    kmeans = KMeans(
        n_clusters=i, init="k-means++", max_iter=300, n_init=10, random_state=42
    )
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

plt.rcParams["figure.figsize"] = (7, 5)
plt.plot(range(1, 11), wcss)
plt.title("K-Means Clustering(The Elbow Method)", fontsize=20)
plt.xlabel("K")
plt.ylabel("WCSS")
plt.suptitle("Fig. 6: The Elbow Method", fontweight="bold", horizontalalignment="right")
plt.show()

# K-means clustering
kmeans = KMeans(n_clusters=4, init="k-means++", max_iter=300, n_init=10, random_state=0)
ymeans = kmeans.fit_predict(X)

plt.rcParams["figure.figsize"] = (10, 10)
plt.title("Cluster of Dataset", fontsize=30)

plt.scatter(X[ymeans == 0, 0], X[ymeans == 0, 1], s=100, c="pink", label="Cluster 0")
plt.scatter(X[ymeans == 1, 0], X[ymeans == 1, 1], s=100, c="orange", label="Cluster 1")
plt.scatter(
    X[ymeans == 2, 0], X[ymeans == 2, 1], s=100, c="lightgreen", label="Cluster 2"
)
plt.scatter(X[ymeans == 3, 0], X[ymeans == 3, 1], s=100, c="red", label="Cluster 3")
plt.scatter(
    kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=50, c="black"
)

plt.xlabel("First Feature")
plt.ylabel("Second Feature")
plt.legend()

plt.suptitle('Fig 10: Clustering Results', fontsize=20, y=1, ha='right')

plt.suptitle("Fig. 7: Clustering Results", fontweight="bold", horizontalalignment="right")
plt.show()
```

### **Equations**

The K-means algorithm's goal is to minimize an objective function that is
The expression
$$J=\sum_{j=1}^k \sum_{i=1}^n ||x_i^{j} - C_j||^2$$

The Euclidean distance between data points $x_i$ and the centroid $C_j$ is
$$||x_i^{j} - C_j||^2$$

By calculating the group mean, we recalculate the new centroids.
$$v_i = \left(\frac{1}{c_i}\right)\sum_{j=1}^{c_i} x_j$$

The code provided demonstrates an example of Principal Component Analysis (PCA) using the scikit-learn library. Here's a summary of what the code accomplishes in a few paragraphs:

- The necessary libraries are imported, including `matplotlib.pyplot` for data visualization, `numpy` for numerical operations, `sklearn.datasets` for loading the Iris dataset, and `sklearn.decomposition.PCA` for performing PCA.

- The Iris dataset is loaded using `datasets.load_iris()`. The dataset contains measurements of sepal length, sepal width, petal length, and petal width for three different species of Iris flowers.

- A covariance matrix is created using `np.corrcoef(iris.data.T)`. The transpose of the data is passed to calculate the correlation coefficients between the features. The resulting covariance matrix represents the relationships between the features.

- The covariance matrix is visualized as a heatmap using `plt.matshow()`. Each cell in the heatmap represents the covariance between two features, with color indicating the strength of the correlation. The color bar next to the heatmap provides a reference for the color scale. Text annotations are added to each cell to display the covariance values.

- PCA is performed on the Iris dataset using `PCA(n_components=2)`. The `n_components` parameter is set to 2, indicating that we want to reduce the dimensionality of the data to two principal components. The `fit_transform()` method is applied to the dataset, which calculates the principal components based on the covariance matrix and transforms the data accordingly.

- The principal components are visualized by creating a scatter plot. Each data point is represented by its values on the first and second principal components, with colors indicating the corresponding Iris species. The x-axis is labeled as "Principal Component 1," the y-axis is labeled as "Principal Component 2," and the title of the plot is set as "PCA of Iris Dataset." A color bar is added to the plot to provide a reference for the color mapping.

- A scree plot is generated to show the explained variance ratio of each principal component. The explained variance ratio represents the proportion of the total variance explained by each principal component. The plot displays the explained variance ratio for each principal component, with the x-axis representing the principal component number and the y-axis representing the explained variance ratio. The scree plot helps determine the number of principal components to retain based on the amount of variance they explain.

In summary, the code loads the Iris dataset, creates a covariance matrix, visualizes the covariance matrix as a heatmap, performs PCA to reduce the data to two principal components, visualizes the principal components in a scatter plot, and generates a scree plot to show the explained variance ratio. PCA is a dimensionality reduction technique commonly used for feature extraction and data visualization.

```python
# Principal Components Analysis (PCA) Example
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.decomposition import PCA

# Load Iris Dataset
iris = datasets.load_iris()

# Create a Covariance Matrix
cov_data = np.corrcoef(iris.data.T)
cov_data

# Plotting the covariance matrix using a heatmap
img = plt.matshow(cov_data, cmap=plt.cm.rainbow)
plt.colorbar(img, ticks=[-1, 0, 1], fraction=0.045)
for x in range(cov_data.shape[0]):
    for y in range(cov_data.shape[1]):
        plt.text(
            x,
            y,
            "%0.2f" % cov_data[x, y],
            size=12,
            color="black",
            ha="center",
            va="center",
        )

plt.show()

# Perform PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(iris.data)

# Plotting the principal components
plt.figure(figsize=(10, 8))
plt.scatter(principal_components[:, 0], principal_components[:, 1], c=iris.target)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Iris Dataset')
plt.colorbar()
plt.show()

# Scree plot
explained_variance = pca.explained_variance_ratio_
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance) + 1), explained_variance, 'o-', linewidth=2, color='blue')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Scree Plot')
plt.show()
```

### **Disadvantages**

1. **Loss of Information**: While PCA reduces dimensionality, it may also discard some information that can be critical. The reduction process involves keeping only the components with the highest variance, which may result in the loss of some important features.

2. **Interpretability**: The principal components are linear combinations of the original features, which can be difficult to interpret. It may be challenging to understand the relationship between the principal components and the original variables.

3. **Assumes Linearity**: PCA assumes that the relationships between variables are linear. It may not capture more complex, non-linear relationships in the data.

### **Equations**

The linear transformation is $$y = WX$$, in which $y$ represents the primary components, $X$ the original features, and $W$ the weights.

A larger percentage of the variability in the original data is explained by the principal components, $y_i$, and the weights $W_{ij}$ indicate how much of our original features went into creating the final principal components.

Let us look at a two-variable example: $$y_i = w_{i,1} x_1 + w_{i2} x_2$$

The weights $(w_{i,1}, w_{i,2})$ are referred to as the **loadings** components. The majority of the variation in our dataset is explained by the first principal component, $y_1$, whereas the majority of the remaining changes are explained by $y_2$, which is orthogonal to $y_1$.


### **Features**

- Variance Preservation: PCA captures the maximum variance in the data with the fewest number of principal components.
- Orthogonality: Principal components are orthogonal, ensuring they capture different dimensions of the data.
- Data Transformation: PCA transforms the data into a new coordinate system where each axis represents a principal component.

### **Guide**

- Inputs: Feature matrix $X$

- Outputs: Principal components, transformed feature matrix $X_{pca}$

### **Hyperparameters**

- Singular value decomposition (SVD) Solver: The algorithm to use for computing the principal components. Options include 'auto', 'full', 'arpack', and 'randomized'.

##**Team Member A: Ridge Regression**

Hyperparameter tuning is an essential step in optimizing the performance of machine learning models, including Ridge regression. Hyperparameters are parameters that are not learned from the data but are set before the learning process begins. They control the behavior and flexibility of the model and can significantly impact its effectiveness.

In the case of Ridge regression, there are a few key hyperparameters that can be tuned to improve the model's performance:

1. **Alpha (α)**: The alpha hyperparameter controls the regularization strength in Ridge regression. It determines the balance between fitting the training data and reducing the magnitude of the coefficients to prevent overfitting. A higher alpha value increases the regularization strength, leading to more shrinkage of the coefficient values. Conversely, a lower alpha value reduces the regularization strength, allowing the model to fit the training data more closely.

2. **Solver**: The solver hyperparameter determines the numerical method used to solve the Ridge regression optimization problem. Different solvers are available, such as 'auto', 'svd', 'cholesky', 'lsqr', and 'saga'. The choice of solver depends on the size of the problem, the desired accuracy, and computational efficiency. The 'auto' option automatically selects the most appropriate solver based on the input data.

3. **Normalize**: The normalize hyperparameter determines whether the input features should be normalized before fitting the Ridge regression model. Normalization scales the features to have zero mean and unit variance, which can be crucial when the input features are on different scales. Setting normalize to True ensures that each feature is scaled, while False leaves the features as they are.

To tune these hyperparameters, various techniques can be employed, such as grid search, random search, or Bayesian optimization. These techniques involve systematically exploring the hyperparameter space and evaluating the model's performance using cross-validation or a separate validation set. The optimal combination of hyperparameters is then selected based on the best performance metric, such as mean squared error (MSE) or R-squared.

##**Team Member B: k-means Clustering**


Hyperparameter tuning is an important step in optimizing the performance of machine learning models, including k-means clustering. Hyperparameters are parameters that are set before the learning process begins and influence the behavior and performance of the model.

In k-means clustering, there are a few key hyperparameters that can be tuned to improve the clustering results:

1. **Number of clusters (k):** The number of clusters to be created is a crucial hyperparameter in k-means clustering. It determines the granularity of the clustering and directly affects the interpretation of the results. Selecting an appropriate value for k is essential to capture meaningful patterns in the data. However, determining the optimal number of clusters is often a challenging task and may require domain knowledge or exploratory analysis.

2. **Initialization method:** The initialization method determines how the initial cluster centroids are chosen in the algorithm. Different initialization methods, such as random initialization, k-means++, or custom initialization, can be used. The choice of initialization method can impact the convergence speed and final clustering results. For example, the k-means++ initialization method tends to produce better initial centroids, leading to faster convergence and potentially better clustering.

3. **Maximum number of iterations:** The maximum number of iterations is a hyperparameter that determines the maximum number of times the k-means algorithm iterates to converge. Each iteration includes reassigning data points to the nearest centroid and updating the centroids. Setting an appropriate maximum number of iterations is important to ensure convergence. If the algorithm reaches the maximum number of iterations without converging, it may indicate that the clusters are not well-separated or that the chosen number of clusters is not suitable for the data.

To tune these hyperparameters, various techniques can be employed. One common approach is to use the elbow method or silhouette analysis to evaluate the clustering performance for different values of k. The elbow method plots the sum of squared distances (inertia) against the number of clusters and identifies the "elbow" point where the rate of decrease in inertia diminishes. Silhouette analysis measures the quality of clustering by calculating the average silhouette coefficient for each data point, indicating how well it belongs to its own cluster compared to other clusters.

Example hyperparameters for k-means clustering:

1. Number of clusters (k): [2, 3, 4, 5]
2. Initialization method: ['random', 'k-means++']
3. Maximum number of iterations: [100, 200, 300]

By systematically trying different values for these hyperparameters and evaluating the clustering performance using appropriate metrics, the optimal set of hyperparameters can be determined.

Hyperparameter tuning is crucial for obtaining meaningful and accurate clusters in k-means clustering. It allows the model to adapt to the characteristics of the data and improve the quality of the clustering results.

##**Team Member C: Principal Components Analysis (PCA)**

Hyperparameter tuning is an essential step in optimizing the performance of machine learning models, including Principal Component Analysis (PCA). Although PCA itself does not have many hyperparameters, there are some important considerations when applying PCA and tuning related parameters.

1. **Number of components:** The number of components is a critical hyperparameter in PCA. It determines the dimensionality reduction and the amount of information retained from the original dataset. Selecting the appropriate number of components is crucial to balance the reduction in dimensionality and the preservation of important information. It can be determined by analyzing the cumulative explained variance ratio or by setting a threshold for the desired level of explained variance.

2. **Scaling or normalization:** PCA is sensitive to the scales of the input features. Therefore, it is often necessary to scale or normalize the features before applying PCA. Scaling ensures that each feature has equal importance in the analysis and prevents features with larger scales from dominating the principal components. Common scaling methods include standardization (mean centering and variance scaling), min-max scaling, or robust scaling.

3. **Covariance matrix type:** PCA can be performed using either the covariance matrix or the correlation matrix of the input data. The choice of covariance matrix type depends on the nature of the data and the desired analysis. Using the covariance matrix considers the variances and covariances of the original variables, while using the correlation matrix scales the variables to have unit variances and focuses on the linear relationships between variables.

4. **Preprocessing techniques:** In some cases, additional preprocessing techniques can be applied before performing PCA to address specific challenges. For example, handling missing values, outlier detection and removal, or feature selection techniques can be utilized to improve the quality of the PCA results. These preprocessing techniques can influence the performance and interpretation of PCA.

To tune these hyperparameters, several approaches can be employed. Cross-validation or validation set-based evaluation can be used to assess the performance of PCA with different hyperparameter settings. For example, the number of components can be varied, and the impact on the reconstruction error or explained variance can be measured. Additionally, domain knowledge and exploratory data analysis can guide the selection of appropriate hyperparameters.

Example hyperparameters for PCA:

1. Number of components: [2, 5, 10]
2. Scaling or normalization: ['standardization', 'min-max scaling']
3. Covariance matrix type: ['covariance', 'correlation']

By systematically trying different values for these hyperparameters and evaluating the performance of PCA on relevant metrics, the optimal configuration can be determined.

Hyperparameter tuning in PCA allows for fine-tuning the dimensionality reduction process and adapting it to the specific requirements of the dataset and analysis objectives, leading to improved results.

Keywords: Principal Component Analysis, PCA, Hyperparameter Tuning, Number of Components, Scaling, Covariance Matrix, Preprocessing Techniques, Cross-validation, Explained Variance Ratio.
