```python
#SCENARIO 4: THIS INVOLVES LENDING OUT A STOCK. WE'LL BE LOOKING AT TESLA STOCKS OVER THE PAST 2 YEARS FOR THIS SCENARIO
import requests
import pandas as pd

api_key = '4JDzcxDvnNGi5GEtaaBJghdK1v3M7emz'

symbol = 'TSLA'

end_date = '2023-09-20'
start_date = '2021-09-20'

endpoint = f'https://api.polygon.io/v2/aggs/ticker/{symbol}/range/1/day/{start_date}/{end_date}?apiKey={api_key}'


response = requests.get(endpoint)

if response.status_code == 200:
    data = response.json()
    

    prices = [item['c'] for item in data['results']]
    timestamps = [pd.Timestamp(item['t'], unit='ms') for item in data['results']]
    
    
    df = pd.DataFrame({'Date': timestamps, 'Price': prices})
    
    
    df['Return'] = df['Price'].pct_change() * 100 
    

    print("Tesla's Stock Data:")
    print(df)
else:
    print(f"Error {response.status_code}: Unable to retrieve data from the API.")
```

```python
import matplotlib.pyplot as plt


df_last_2_years = df[df['Date'] >= df['Date'].max() - pd.DateOffset(years=2)]

plt.figure(figsize=(12, 6))
plt.bar(df_last_2_years['Date'], df_last_2_years['Return'], width=20, color='blue', alpha=0.7)
plt.xlabel("Year")
plt.ylabel("Return (%)")
plt.title("Tesla's Stock Returns Over the Last 2 Years")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

plt.show()
```

```python
#Scenario 5 deals with lending a publicly traded bond. We'll look at the 10-Year Treasury Constant Maturity Rate, which is the bond series for the 10 year U.S Treasury Bond

import pandas as pd
from fredapi import Fred
import numpy as np

api_key = '79c8d218507d76db41969c62b12ba88f '

fred = Fred(api_key=api_key)


treasury_series_id = 'GS10'

end_date = pd.to_datetime('today')
start_date = end_date - pd.DateOffset(years=10)

yield_data = fred.get_series(treasury_series_id, start_date, end_date)


returns = yield_data.resample('Y').last().pct_change().dropna()


result_df = pd.DataFrame(columns=['Year', 'Annual Yield', 'Annual Returns', 'Volatility'])

for year in range(start_date.year, end_date.year + 1):
    year_start = pd.Timestamp(year, 1, 1)
    year_end = pd.Timestamp(year, 12, 31)
    
    year_yield_data = yield_data[(yield_data.index >= year_start) & (yield_data.index <= year_end)]
    annual_yield = year_yield_data.iloc[-1]
    
    if year > start_date.year:
        annual_returns = (year_yield_data.iloc[-1] / year_yield_data.iloc[0] - 1) * 100
    else:
        annual_returns = 0.0
    
    if len(year_yield_data) > 1:
        annual_volatility = np.std(year_yield_data.pct_change().dropna()) * np.sqrt(252)  # Annualized volatility
    else:
        annual_volatility = 0.0
    
    result_df = result_df.append({
        'Year': year,
        'Annual Yield': annual_yield,
        'Annual Returns': annual_returns,
        'Volatility': annual_volatility,
    }, ignore_index=True)

print(result_df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(result_df['Year'], result_df['Volatility'], marker='o', linestyle='-', color='blue')
plt.title('Volatility of U.S. Treasury Bond Over the Past 10 Years')
plt.xlabel('Year')
plt.ylabel('Annual Volatility')
plt.grid(True)
plt.show()
```
