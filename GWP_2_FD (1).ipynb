## Section 1: Concepts

**Statistical Related Risk: Volatility**

- Volatility measures the variability or dispersion of returns for a security or market index over a specific period.
- It quantifies the level of uncertainty and risk associated with an investment.
- Higher volatility suggests that the price of the asset can change significantly in a short amount of time, which can lead to both potential gains and losses.
- Investors often consider volatility when assessing risk because it can indicate the potential for large losses or gains. For example, a highly volatile stock might be riskier than one with lower volatility because it can experience rapid and unpredictable price movements.
- Volatility is commonly calculated using the standard deviation of historical returns. The larger the standard deviation, the greater the volatility.

**Statistical Related Risk: Correlation**

- Correlation measures the statistical relationship between the returns of two assets or variables.
- It helps investors understand how two assets move in relation to each other.
    - A correlation coefficient can range from -1 to +1:
    - A correlation of +1 implies a perfect positive correlation, meaning that the two assets move in the same direction.
    - A correlation of -1 implies a perfect negative correlation, meaning that the two assets move in opposite directions.
    - A correlation of 0 means there is no linear relationship between the two assets.
- Correlation is vital in portfolio management because it can be used to diversify risk. Diversification involves investing in assets that are not highly correlated with each other to reduce the overall risk of a portfolio.
- When assets in a portfolio are negatively correlated or uncorrelated, their price movements tend to offset each other. This reduces the portfolio's overall risk because losses in one asset may be offset by gains in another.

## Section 2: Real-World Examples

In this section, we'll explore real-world examples of volatility and correlation using two asset classes: 

- Asset Class 1: S&P 500 Index (US Equities)
- Asset Class 2: Single Stock Large US Cap (IBM)

We will analyze historical data of the S&P 500 index from January 1993 to December 2023, obtained from Yahoo Finance.

#### Volatility

To calculate the volatility of the S&P 500 index, we'll follow these steps:

1. Calculate the daily returns of the S&P 500 index.
2. Calculate the standard deviation of the daily returns, which represents the volatility.

Here's the code to calculate the volatility:


```python
# Import dependencies
import pandas as pd
import numpy as np
import sys
# !{sys.executable} -m pip install yfinance
import yfinance


# Load historical S&P 500 data 
sp500_data = yfinance.download (tickers = "SPY", start = "1993-01-01",
                              end = "2023-04-01", interval = "1d")



# Calculate daily returns
sp500_data['Daily Returns'] = sp500_data['Adj Close'].pct_change()

#Read 10 row
sp500_data.head(10)
```

```python
# Calculate volatility (standard deviation of daily returns)
volatility = np.std(sp500_data['Daily Returns'])

print('Volatility of S&P 500 index: ', round(volatility, 5))

#The volatility of the S&P 500 index over this period was 0.01191.
```

**Correlation**

To calculate the correlation between the S&P 500 index and another asset (IBM), we can calculate the daily returns of both assets and then calculate the correlation between them. Here, we will calculate the correlation between the S&P 500 index and IBM

```python
# Load data for IBM
IBM = yfinance.download (tickers = "IBM", start = "1993-01-01",
                              end = "2023-04-01", interval = "1d")

IBM['Returns'] = IBM['Adj Close'].pct_change()
IBM.head(10)
```

```python
# Calculate volatility
volatility = IBM['Returns'].std()
print('Volatility of IBM: ', round(volatility, 5))

#The volatility of IBM:  0.01753
```

```python
# Merge S&P 500 and IBM data
data = pd.concat([sp500_data['Daily Returns'], IBM['Returns']], axis=1, join='inner')
data.columns = ['SP500 Returns', 'IBM Returns']

# Calculate correlation
correlation = data['SP500 Returns'].corr(data['IBM Returns'])
print('Correlation between S&P 500 and IBM: ', round(correlation, 5))
```

We illustrate the presence and absence of correlation, and hence calculate the correlation between daily returns of two different asset classes - one with high correlation and one with low correlation.

We can use Coca Cola and Pepsi stocks as example of assets with high correlation and the price of US T_bill and the price of Amazon as an example of assets with low correlation.


```python
#load Coca Cola
Coke = yfinance.download (tickers = "Coke", start = "2003-01-01",
                              end = "2023-04-01", interval = "1d")
```

```python
Coke['Returns'] = Coke['Adj Close'].pct_change()
```

```python
#load Pepsi
PEP = yfinance.download (tickers = "PEP", start = "2003-01-01",
                              end = "2023-04-01", interval = "1d")

PEP['Returns'] = PEP['Adj Close'].pct_change()
```

```python
# Merge C and PEP data
high_data = pd.concat([Coke['Returns'], PEP['Returns']], axis=1, join='inner')
high_data.columns = ['Coke Returns', 'PEP Returns']
```

```python
#load US T Bill
T_BIL = yfinance.download (tickers = "TBIL", start = "2003-01-01",
                              end = "2023-04-01", interval = "1d")

T_BIL['Returns'] = T_BIL['Adj Close'].pct_change()
```

```python
#load AMZN
AMZN = yfinance.download (tickers = "AMZN", start = "2003-01-01",
                              end = "2023-04-01", interval = "1d")

AMZN['Returns'] = AMZN['Adj Close'].pct_change()

```

```python
# Merge TBIL and AMZN data
low_data = pd.concat([T_BIL['Returns'], AMZN['Returns']], axis=1, join='inner')
low_data.columns = ['TBIL Returns', 'AMZN Returns']
```

```python
high_data.head(10),low_data.head(10)
```

```python
# Calculate correlation matrix of cola cola and pepsi, and 
cor_matrix = pd.concat([high_data, low_data], axis=1).corr()

# Extract relevant correlations
high_cor = cor_matrix.iloc[0:2, 0:2]
low_cor = cor_matrix.iloc[2:4, 2:4]

#plot the correlation matrices:

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,7), sharex=True, sharey=True)

im1 = ax1.imshow(high_corr, cmap='coolwarm', vmin=-1, vmax=1)
ax1.set_title('Coca Cola and Pepsi Stock Returns Correlation')
ax1.set_xticks([0,1])
ax1.set_yticks([0,1])

ax1.set_xticklabels(['Coca Cola', 'Pepsi'])

# fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20,5))

im2 = ax2.imshow(low_corr, cmap='coolwarm', vmin=-1, vmax=1)
ax2.set_title('T Bill and Amazon Stock Returns Correlation')
ax2.set_xticks([0,1])
ax2.set_yticks([0,1])
ax2.set_xticklabels(['T Bill', 'Amazon'])
cbar = fig.colorbar(im2, ax=[ax1, ax2], orientation='vertical')
```

```python
# combined data frame
df_combined = pd.concat([AMZN['Adj Close'], T_BIL['Adj Close']], axis=1, join='inner')
df_combined.columns = ['AMZN', 'T_BIL']

# calculate rolling correlation with a window of 30 days
rolling_corr = df_combined['AMZN'].rolling(window=30).corr(df_combined['T_BIL'])

# plot the rolling correlation
fig, ax = plt.subplots(figsize=(20, 8))
rolling_corr.plot(ax=ax, title='Rolling Correlation (30-day window)')
plt.show()
```
